# Sentiment Analysis

Opinion mining (sometimes known as sentiment analysis or emotion AI) refers to the use of natural language processing, text analysis, computational linguistics, and biometrics to systematically identify, extract, quantify, and study affective states and subjective information. Sentiment analysis is widely applied to voice of the customer materials such as reviews and survey responses, online and social media, and healthcare materials for applications that range from marketing to customer service to clinical medicine. (Wikipedia)

## 1. Data Analysis

```python
# Importing the libraries
import os
import glob
import pickle
from wordcloud import WordCloud, STOPWORDS
from sklearn.utils import shuffle
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Loading the dataset
def read_imdb_data(data_dir = "./data/imdb-reviews/"):
    """
    Loading the IMDB dataset.
    
    Parameters
    ----------
    - data_dir : Directory of dataset
    
    Return
    ------
    - data : The review text
    - label : The label of text; positive or negative
    
    """
    # Initializing a dictionary for data and labels
    data = {}
    labels = {}
    
    # Iterating through the [training set, test set] 
    for data_type in ['train', 'test']:
        
        # Initializing an empty dictionary for the train and test
        data[data_type] = {}
        labels[data_type] = {}
        
        # Iterating through [positive, negative]
        for sentiment in ['pos', 'neg']:
            
            # Initializing an empty list for each sentiment
            data[data_type][sentiment] = []
            labels[data_type][sentiment] = []
            
            # Getting the list of files
            path = os.path.join(data_dir, data_type, sentiment, "*.txt") # Joinig the path name
            files = glob.glob(path) # Getting the name of files
                
            # Iterating through file names
            for f in files:
                
                # Opening the text
                with open(f) as review:
                
                    # Assigning values to our dictionary
                    data[data_type][sentiment].append(review.read())
                    labels[data_type][sentiment].append(sentiment)
                
            # Checking the length of data and labels in order to be equal
                assert (len(data[data_type][sentiment]) == len(labels[data_type][sentiment])), "Data and label sizes does not match"
                
    return data, labels
```

```python
# Loading the files
data, labels = read_imdb_data()
```

##### Form Training and Test set

```python
def prepare_imdb_data(data):
    """
    Converting the dataset into the training set and test set.
    
    Parameters
    ----------
    - data : the dataset we want to split.
    
    Return
    ------
    - data_train : Training set
    - data_test : Test set
    - labels_train : Labels for the training set
    - labels_test : Labels for the test set
    
    """
    # Combining the pos and neg reviews
    data_train = data['train']['pos'] + data['train']['neg']
    data_test = data['test']['pos'] + data['test']['neg']
    labels_train = labels['train']['pos'] + labels['train']['neg']
    labels_test = labels['test']['pos'] + labels['test']['neg']
    
    # Shuffling
    data_train, labels_train = shuffle(data_train, labels_train)
    data_test, labels_test = shuffle(data_test, labels_test)
    
    return data_train, data_test, labels_train, labels_test
```

```python
data_train, data_test, labels_train, labels_test = prepare_imdb_data(data)

print("IMDb dataset: training set = {}, test set = {}".format(len(data_train), len(data_test)))
```

## 2. Preprocessing

```python
# Importing the libraries
from bs4 import BeautifulSoup # For removing HTML tags
import re # RegEx for removing non-letter characters
import nltk
from nltk.stem.porter import *
from nltk.corpus import stopwords

# Downloading the stopwords
nltk.download("stopwords")

# Initializing the stemmer
stemmer = PorterStemmer()
```

```python
# Converting the reviews into preprocessed words

def review_to_words(review):
    """
    Converting the reviews into preprocessed words.
    
    Parameters
    ----------
    - review : The raw text for the review.
    
    Return 
    ------
    words : Preprocessed words.
    
    """
    # Eliminating the HTML tags
    text = BeautifulSoup(review, "html5lib").get_text()
    
    # Eliminating the non-letters characters
    text = re.sub(r"[^a-zA-Z0-9]", " ", text.lower())
    
    # Splitting text into words
    words = text.split()
    
    # Eliminating the stopwords
    words = [w for w in words if w not in stopwords.words("english")]
    
    # Stemming the words
    words = [PorterStemmer().stem(w) for w in words]
    
    return words
```

```python
# Checking the review_to_words
review_to_words("""This is just a <em>test</em>.<br/><br/> But if it wasn't a test, it would make for a <b>Great</b> movie review!""")
```

```python
# Cache file name for storing the file
cache_dir = os.path.join("cache", "sentiment_analysis")

# Making the directory
os.makedirs(cache_dir, exist_ok = True)

def preprocess_data(data_train, data_test, labels_train, labels_test, cache_dir = cache_dir, cache_file = "preprocessed_data.pkl"):
    """
    Converting each review to words; Reading from cache if available.
    
    Parameters
    ----------
    - data_train : Training set
    - data_test : Test set
    - labels_train : Labels for the training set
    - labels_test : Labels for the test set
    - cache_dir : Directory of cache
    - cache_file : Cache file
    
    Return
    ------
    - words_train
    - words_test
    - labels_train 
    - labels_test

    """
    # Initializing the cache file
    cache_data = None
    
    # 1. Read cache data; if it's not None
    if cache_file is not None:
        
        try:
            # Path name of cach files
            file_name = os.path.join(cache_dir, cache_file)
            
            # Opening and loading the cache file
            with open(filename, 'rb') as f:
                cache_data = pickle.load(f)
            print("Read the preprocessed data from cache file: ", cache_file)
            
        # If unable to read a cach file
        except:
            pass
        
    # 2. Heavy lifting; if cache data is None
    if cache_data is None:
        
        # Preprocessing the training and test set
        words_train = list(map(review_to_words, data_train))
        words_test = list(map(review_to_words, data_test))
        
        # Write to cache file
        if cache_file is not None:
            
            # Getting the cache data
            cache_data = dict(words_train = words_train,
                              words_test = words_test,
                              labels_train = labels_train,
                              labels_test = labels_test)
            
            # Writing the preprocessed data into cache file
            file_name = os.path.join(cache_dir, cache_file)
            with open(file_name, "wb") as f:
                pickle.dump(cache_data, f)
            print("Wrote preprocessed data into cache file: ", cache_file)
        
    else:
        # Unpacking the data loaded from cache file
        words_train, words_test, labels_train, labels_test = (cache_data['words_train'], 
                                                              cache_data['words_test'], 
                                                              cache_data['labels_train'], 
                                                              cache_data['labels_test'])
        
    return words_train, words_test, labels_train, labels_test
```

```python
# Preprocessing the data
words_train, words_test, labels_train, labels_test = preprocess_data(data_train,
                                                                     data_test,
                                                                     labels_train,
                                                                     labels_test)
```

## 3. Extracting the Bag-of-Words Features

```python
# Importing the libraries
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.externals import joblib # An enhanced version of pickle that is more efficient for storing NumPy arrays
import sklearn.preprocessing as pr
```

```python
def extract_BoW_features(words_train, words_test, vocabulary_size = 5000, cache_dir = cache_dir, cache_file = "bow_features.pkl"):
    """
    Extracting Bag-of-Words for a given set of documents that already preprocessed into words.
    
    Parameters
    ----------
    words_train : Preprocessed training set
    words_test : Preprocessed test set
    vocabulary_size : Size of vocabulary
    cache_dir : Directory of caches
    cache_file : Name of the cache
    
    Return
    ------
    features_train : BoW features for training set
    features_test : BoW features for test set
    vocabulary : Vocabulary
    
    """
    # Initializing the cache data
    cache_data = None
    
    # 1. Read cache data; if it's not None
    if cache_file is not None:
        
        try:
            # Path name of the cache file
            file_name = os.path.join(cache_dir, cache_file)
            
            # opening and loading the cache file
            with open(file_name, "rb") as f:
                joblib.load(f)
                print("Read files from cache file.")
               
        # If unable to open the cache file
        except:
            pass
        
    # 2. Heavy lifting; if cache data is None
    if cache_data is None:
        
        # Initializing a vectorizer
        vectorizer = CountVectorizer(max_features = vocabulary_size,
                                     preprocessor = lambda x: x, # Already preprocessed
                                     tokenizer = lambda x: x) # Already tokenized
        
        # Fiting and transforming the vectorizer to training document
        features_train = vectorizer.fit_transform(words_train).toarray()
        
        # Transforming the vectorizer to test document
        features_test = vectorizer.transform(words_test).toarray()
        
        # Writing to cache file
        if cache_file is not None:
            
            # Getting the vocabulary
            vocabulary = vectorizer.vocabulary_
            
            # Getting the cache data
            cache_data = dict(features_train = features_train,
                              features_test = features_test,
                              vocabulary = vocabulary)
            
            # Writing into cache file
            file_name = os.path.join(cache_dir, cache_file)
            with open(file_name, "wb") as f:
                joblib.dump(cache_data, f)
            print("Wrote features to cache file: ", cache_file)
                
    else:
        # Unpacking the data loaded from cache file
        features_train, features_test, vocabulary = (cache_data['features_train'], 
                                                     cache_data['features_test'],
                                                     cache_data['vocabulary'])
        
    return features_train, features_test, vocabulary
```

```python
# Bag of Words features for training and test set
features_train, features_test, vocabulary = extract_BoW_features(words_train, words_test)
```

### Zipf's law

[Zipf's law](https://en.wikipedia.org/wiki/Zipf%27s_law), named after the famous American linguist George Zipf, is an empirical law stating that given a large collection of documents, the frequency of any word is inversely proportional to its rank in the frequency table. So the most frequent word will occur about twice as often as the second most frequent word, three times as often as the third most frequent word, and so on. In the figure below we plot number of appearances of each word in our training set against its rank.

```python
# Number of occurance of words in the training set
word_freq = features_train.sum(axis=0)

# Sort (descending order) the word_freq
sorted_word_freq = np.sort(word_freq)[::-1]

# Ploting
plt.plot(sorted_word_freq)
plt.gca().set_xscale('log')
plt.gca().set_yscale('log')
plt.xlabel('Rank')
plt.ylabel('Number of occurrences')
plt.show()
```

##### Gradient Boosting Classifier

```python
# Number of estimator (Hyperparameters)
n_estimators = 32

# Gradient Boosting Classifier
def classify_gboost(X_train, X_test, y_train, y_test):        
    
    # Initialize classifier
    classifier_2 = GradientBoostingClassifier(learning_rate = 1.0, 
                                     n_estimators = n_estimators,
                                     max_depth = 1, 
                                     random_state = 0)

    # Fitting the model to training set and test set
    classifier_2.fit(X_train, y_train)
    
    print("[{}] Accuracy: \ntrain = {}, test = {}".format(classifier_2.__class__.__name__,
                                                        classifier_2.score(X_train, y_train),
                                                        classifier_2.score(X_test, y_test)))
    
    return classifier_2
```

```python
classifier_2 = classify_gboost(features_train, features_test, labels_train, labels_test)
```

## 5. Using RNNs

```python
# Importing the libraries
from keras.datasets import imdb
from keras.preprocessing import sequence # Pads sequences to the same length.
from keras.models import Sequential
from keras.layers import Embedding, LSTM, Dense, Dropout
```

```python
# Vocabulary size
vocabulary_size = 5000

# Splitting the dataset into Training and Test set
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words = vocabulary_size)
```

##### Pad Sequences

In order to feed this data into your RNN, all input documents must have the same length. Let's limit the maximum review length to `max_words` by truncating longer reviews and padding shorter reviews with a null value (0). You can accomplish this easily using the [`pad_sequences()`](https://keras.io/preprocessing/sequence/#pad_sequences) function in Keras. For now, set `max_words` to 500.

```python
# Maximum number of words per document
max_words = 500

# Padding the sequences in X_train and X_test
X_train = sequence.pad_sequences(X_train, maxlen = max_words)
X_test = sequence.pad_sequences(X_test, maxlen = max_words)
```

##### Design an RNN model for sentiment analysis

Build your model architecture in the code cell below. We have imported some layers from Keras that you might need but feel free to use any other layers / transformations you like.

Remember that your input is a sequence of words (technically, integer word IDs) of maximum length = `max_words`, and your output is a binary sentiment label (0 or 1).

```python
# Embedding size
embedding_size = 32

# Designing the model
model = Sequential()
model.add(Embedding(input_dim = vocabulary_size,
                    output_dim = embedding_size,
                    input_length = max_words))
model.add(LSTM(units = 100))
model.add(Dense(units = 1, activation = "sigmoid"))

print(model.summary())
```

##### Training and evaluating the model

```python
# Compiling the model
model.compile(optimizer = "adam", loss = "binary_crossentropy", metrics = ["accuracy"])
```

```python
# Hyperparameters
batch_size = 64
num_epochs = 3
```

```python
# Splitting the training set into training set and validation set
X_valid, y_valid = X_train[:batch_size], y_train[:batch_size]
X_train_2, y_train_2 = X_train[batch_size:], y_train[batch_size:]
```

```python
# Training the model
model.fit(x = X_train_2, 
          y = y_train_2, 
          batch_size = batch_size,
          epochs = num_epochs, 
          validation_data = (X_valid, y_valid))
```

```python
# Evaluating the model on test saet
scores = model.evaluate(x = X_test, 
                       y = y_test, 
                       verbose = 0)
print("Test set accuracy: ", scores[1]) # scores[1] should correspond to accuracy if you passed in metrics=['accuracy']
```
